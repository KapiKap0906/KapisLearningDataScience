Both loc and iloc are row first indexing, the opposite of traditional python. But negative indexing also works.

```python
powerlift.iloc[[0, 2, 21], 3]
```

Indexing using numbers will get difficult with scale, you can index using column names instead.

```python
powerlift = pd.read_csv('openpowerlifting-2024-01-06-4c732975.csv')
```

Be careful with loc, it needs an index column to work from what I've seen. loc is used for lablel based selection. iloc is like loc but ignores the dataset's indices so it is simpler to handle.

```python
powerlift.loc[0,'Age']
```

```python
powerlift.loc[:, ['Name', 'BodyweightKg', 'WeightClassKg']]
```

One other small difference between loc and iloc is that iloc uses the python stdlib indexing where 0:3 = [0,1,2] but loc uses the more typical method where 0:3 = [0,1,2,3].

Let's say we want to find the max done any the people in each weight class, this is how we'd do it.

```python
powerlift.groupby('AgeClass').TotalKg.max()
```

We get an output as a Series so we can also further use the .apply() method.

We can further finetune the grouping by grouping mutiple fields at the same time

```python
mi = powerlift.groupby(['AgeClass', 'WeightClassKg']).TotalKg.agg([max])
#.apply(lambda df: df.loc[df.TotalKg.idxmax()])
mi
```

The above table has multiple indexes and is called a multi-index

```python
type(mi.index)
```

```python
mi.reset_index()
```

```python
powerlift.groupby(['MeetCountry']).TotalKg.agg([len, min, max])
```
